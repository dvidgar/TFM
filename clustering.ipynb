# CLUSTERING

### sin casa

```python
import pandas as pd
import glob
from sklearn.cluster import KMeans

path = './limpio' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=0, header=0).dropna(axis='columns')
    li.append(df)

li[0]
```

```python
concatenated = pd.concat(li, join='inner', sort=True)
kmeans = KMeans(n_clusters=5).fit(concatenated)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
concatenated
```

```python
df_centroids = pd.DataFrame(centroids, columns=concatenated.columns.values)
df_centroids.index.names = ['cluster']
df_centroids
```

### con casa 

```python
import pandas as pd
import glob
from sklearn.cluster import KMeans

path = './limpio' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=0, header=0).dropna(axis='columns')
    df['casa'] = filename[-5:-4]
    li.append(df)
```

```python
li[4]
```

# pruebas nuevas

```python
temp = []
for i in range(len(li)):
# concatenated = pd.concat(li, join='inner', sort=True)
    kmeans = KMeans(n_clusters=1).fit(li[i])
    centroids = kmeans.cluster_centers_
    temp.append(pd.DataFrame(centroids, columns = li[i].columns))
```
